```python
import sys
import logging
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
from pathlib import Path

# User defined modules
E9path = Path("C:/", "Users", "ken92", "Documents", "Studies", "E5", "simulation", "E9_simulations")
if str(E9path) not in sys.path:
    sys.path.insert(1, str(E9path))
from E9_fn import util
import E9_fn.E9_models as E9M
```

```python
cl_sys = "#0144a9"
cl_f_sys = "#1F3455"
cl_res = "#d36a00"
cl_f_res = "#906031"
```

```python
fig, (ax_sys, ax_toy) = plt.subplots(1, 2, figsize = (9, 3))

E_range = np.linspace(-1, 1, 201)
mu_sys = 0.4
f_0 = util.fermi_stat(E_range, tau = 0.1, mu = 0)
f_mu = util.fermi_stat(E_range, tau = 0.1, mu = mu_sys)
ax_sys.plot(E_range, np.ones_like(E_range), color = cl_sys, label = r"$\rho_S$")
ax_sys.fill_between(E_range, f_0, color = cl_f_sys, alpha = 0.5)#, label = r"$f(E; \mu = 0)$")
ax_sys.fill_between(E_range, f_mu, color = cl_f_sys, alpha = 0.5, label = r"$f(E; \mu)$")
mu_sys_arrow = FancyArrowPatch((0, 0.5), (mu_sys, 0.5),
                               mutation_scale = 12,
                               linewidth = 0.8,
                               facecolor = "red",
                               shrinkA = 5,
                               shrinkB = 5)
ax_sys.add_patch(mu_sys_arrow)
ax_sys.text(mu_sys / 2, 0.55, r"$\mu$", ha = "center", va = "bottom")
ax_sys.legend()
ax_sys.set_xlim(-1, 1)
ax_sys.set_ylim(0, 3)
ax_sys.set_xlabel(r"$E$", loc = "right", size = 14, labelpad = -15)
ax_sys.set_ylabel(r"$\rho(E)$", y = 0.9, size = 14, labelpad = -5, rotation = "horizontal")
ax_sys.set_xticks([0], [r"$0$"], size = 14)
ax_sys.set_yticks([1], [r"$\frac{L_S}{W}$"], size = 18)
ax_sys.spines[['top', 'right']].set_visible(False)

ax_toy.plot(E_range, np.ones_like(E_range), color = cl_sys, label = r"$\rho_S$")
ax_toy.fill_between(E_range, f_0, color = cl_f_sys, alpha = 0.5)
deltafn_arrow = FancyArrowPatch((0, -0.2), (0, 3),
                               mutation_scale = 12,
                               linewidth = 0.,
                               facecolor = cl_res)
ax_toy.add_patch(deltafn_arrow)
ax_toy.plot([0, 0], [0, 1.5], color = cl_f_res, linewidth = 4, alpha = 0.5)
# ax_toy.legend()
ax_toy.set_xlim(-1, 1)
ax_toy.set_ylim(0, 3)
ax_toy.set_xlabel(r"$E$", loc = "right", size = 14, labelpad = -15)
ax_toy.set_ylabel(r"$\rho(E)$", y = 0.9, size = 14, labelpad = -5, rotation = "horizontal")
ax_toy.set_xticks([0], [r"$0$"], size = 14)
ax_toy.set_yticks([1], [r"$\frac{L_S}{W}$"], size = 18)
ax_toy.spines[['top', 'right']].set_visible(False)

thesis_fig_path = Path(r"C:\Users\ken92\Documents\Studies\E5\Theses\figures\flatband")
band_struc_full_path = Path(thesis_fig_path, "toy_model_schematics.pdf")
fig.savefig(band_struc_full_path, facecolor = "none", dpi = 300, bbox_inches = 'tight')
```
