```python
import util_thm
import glob
import matplotlib.pyplot as plt
import numpy as np
```

# Load data
All the data are combined into one big dataframe.

```python
eos32 = util_thm.read_eos_files("test_chi32")
eos64 = util_thm.read_eos_files("test_chi64")
eos32
```

The parameters in `idx_list` is considered as system parameters, and the ones not in it are observables.

```python
idx_list = ["mu", "beta", "L", "U", "V", "t", "tp", "shift", "int_range"]
eos32 = eos32.set_index(idx_list).sort_index()
eos64 = eos64.set_index(idx_list).sort_index()
eos32
```

For now let's look at one value of `U` at a time.

```python
eos32 = eos32[eos32.index.get_level_values("U") == 3]
eos64 = eos64[eos64.index.get_level_values("U") == 3]
```

Get some numpy arrays that's easier to work with.

```python
mu_vals32  = eos32.index.get_level_values("mu").unique().to_numpy()
beta_vals32 = eos32.index.get_level_values("beta").unique().to_numpy()
N_grid32 = eos32["ntot"].unstack("beta").loc[mu_vals32, beta_vals32].to_numpy()
S_grid32 = eos32["S2"].unstack("beta").loc[mu_vals32, beta_vals32].to_numpy()

mu_vals64  = eos64.index.get_level_values("mu").unique().to_numpy()
beta_vals64 = eos64.index.get_level_values("beta").unique().to_numpy()
N_grid64 = eos64["ntot"].unstack("beta").loc[mu_vals64, beta_vals64].to_numpy()
S_grid64 = eos64["S2"].unstack("beta").loc[mu_vals64, beta_vals64].to_numpy()
```

```python
print(mu_vals32)
if not np.all(mu_vals32 == mu_vals64):
    raise(ValueError("Oh you're not gonna like this"))
```

```python
print(beta_vals32)
if not np.all(beta_vals32 == beta_vals64):
    raise(ValueError("Oh you're not gonna like this"))
```
