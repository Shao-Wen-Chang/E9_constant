```python
import sys
import logging
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
from pathlib import Path

# User defined modules
E9path = Path("C:/", "Users", "ken92", "Documents", "Studies", "E5", "simulation", "E9_simulations")
if str(E9path) not in sys.path:
    sys.path.insert(1, str(E9path))
from E9_fn import util
import E9_fn.E9_models as E9M

plt.style.use(Path(E9path, r'E9_fn/mplstyles/swc.mplstyle'))
```

```python
cl_sys = "#0144a9"
cl_f_sys = "#1F3455"
cl_res = "#d36a00"
cl_f_res = "#906031"
```

```python
fig, ax = plt.subplots(1, 1, figsize = (0.5, 1.5))

deltafn_arrow = FancyArrowPatch((0, -0.2), (0, 3),
                               mutation_scale = 12,
                               linewidth = 0.,
                               facecolor = cl_res)
ax.add_patch(deltafn_arrow)
ax.plot([0, 0], [0, 1.5], color = cl_f_res, linewidth = 4, alpha = 0.5)
ax.set_xlim(-1, 1)
ax.set_ylim(0, 3)
ax.axis("off")
```

```python
fig_s, ax_s = plt.subplots(1, 1, figsize = (2, 1))
taus = [0.05, 0.1, 0.2]
cls = ["#4444ff", "#9944aa", "#ff4444"]

E_range = np.linspace(-1, 1, 201)
ax_s.plot(E_range, np.ones_like(E_range), color = "black")
for tau, cl in zip(taus, cls):
    f_0 = util.fermi_stat(E_range, tau = tau, mu = 0)
    ax_s.plot(E_range, f_0, color = cl)
    ax_s.fill_between(E_range, f_0, color = cl, alpha = 0.2)
    ax_s.set_xlim(-1, 1)
    ax_s.set_ylim(0, 1.2)
    ax_s.set_xticks([0], [""], size = 14)
    # ax_s.set_yticks([1], [r"$\frac{L_S}{W}$"], size = 18)
    ax_s.set_yticks([])
    ax_s.spines[['top', 'right']].set_visible(False)
```
