# Constant $N_{tot}$ and $S_{tot}$
Now I have a three-dimensional parameter space $(n_s, N_{tot}, S_{tot})$ for each interaction strength $V$. At each point in this parameter space, I can evaluate things like $s_s$, $V_{offset}$ and so on, possibly with interpolation. A constant $N_{tot}$ and $S_{tot}$ slice gives these parameters as a function of $n_s$ only.

It should be possible to invert $n_s$ and $V_{offset}$ if one wants to, and we know that this has to be a monotonic function. We can do this inversion after the constant $N, S$ slice.

## Some post processing
### Flat band peculiarities
The flat band must have enough entropy capacity for $\beta$ to go to infinity, at which the total entropy is explicitly given by

$S_{tot}^\circ \equiv S_{tot}|_{\beta = \infty} =
 L \sum_{x = r, s} [-(1 - n_{x,f}) \ln(1 - n_{x,f}) - n_{x,f} \ln n_{x,f}]$

where

$n_{x, f} = \max(2 n_{x,r} - 1, 0)$

is the filling in flat band in the system / reservoir. So things are actually rather fine tuned.


```python
def S_flat(n, L):
    n_f = np.maximum(0., 2 * n - 1.)     # density in the flat band at T = 0
    S = -L * ((1 - n_f) * np.log(1 - n_f) + n_f * np.log(n_f))
    return np.where(np.isnan(S), 0., S)
```

For the case of fixed $N_{tot}$ we look at here, $S_{tot}^\circ$ is only a function of $n_s$.

```python
def S_zeroT(n_s, L, Ntot):
    n = Ntot / 4 / L
    n_r = 2 * n - n_s
    return S_flat(n_s, L) + S_flat(n_r, L)
```

```python
n_s_all = np.linspace(0.05, 0.95, 401)
n_r_all = (2 * ntot - n_s_all)
S_zeroT_all = S_zeroT(n_s_all, L_subregion, Ntot)
S_excess = np.maximum(S_zeroT_all - Stot, 0.)

fig_S, (ax_S, ax_S_excess) = plt.subplots(2, 1, height_ratios = (7, 1))
ax_S.plot(n_s_all, S_flat(n_s_all, L_subregion), ls = "--", label = r"$n_s$")
ax_S.plot(n_s_all, S_flat(n_r_all, L_subregion), ls = "--", label = r"$n_r$")
ax_S.plot(n_s_all, S_zeroT_all, alpha = 0.7, label = "sum")
ax_S.plot(n_s_all, np.full_like(n_s_all, Stot), ls = ":", color = "k", label = r"$S_{tot}$")
ax_S.axvline(n_s_halfnr, color = "r", ls = "--", label = r"$n_{r,f} = 1/2$")
ax_S.set_xticks([])
ax_S.set_ylabel(r"$S_{tot}^\circ$")
ryax = ax_S.secondary_yaxis('right', functions = (lambda x: x / Ntot, lambda x: x * Ntot))
ryax.set_ylabel(r"$\sigma$")
ax_S.legend()
util.add_shades(ax_S, S_excess / np.max(S_excess), n_s_all, alpha = 0.5, color = "pink")

ax_S_excess.plot(n_s_all, S_excess)
ax_S_excess.axvline(n_s_halfnr, color = "r", ls = "--")
ax_S_excess.set_xlabel(r"$n_s$")
ax_S_excess.set_ylabel(r"$S_{tot}^\circ - S_{tot}$")
ryax = ax_S_excess.secondary_xaxis('top', functions = (lambda x: 2 * ntot - x, lambda x: 2 * ntot - x))
ryax.set_xlabel(r"$n_r$")

fig_S.suptitle(rf"Entropy at T = 0 ($N_{{tot}} = {Ntot}$, $S_{{tot}} = {Stot}$; $n = {ntot}$, $\sigma = {sigma2tot:.4f}$)")
fig_S.tight_layout()
if bool_for_thesis and bool_save_fig:
    s_circ_path = Path(thesis_fig_path, f"S_circ_{str_Scase}_{str_Ncase}.pdf")
    # fig_S.savefig(s_circ_path, facecolor = "none", dpi = 300)
```

### Finding the minimum $s_s$
We ask: if we don't ask for a specific $n_s$, what is the $V_{offset}$ that gives us the lowest $s_s$?

It turns out that this is sometimes achieved at very large $n_s$, which is a boring case, because it is just close to a band insulator. Therefore I put an upper limit on $n_s$ .

```python
argmin_sigma2_s_allV = []
for s2_s, n_s in zip(sigma2_s_allV, n_s_allV):
    if np.all(np.isnan(s2_s)):
        argmin_sigma2_s_allV.append(np.array(0))
        continue
    argmin_sigma2_s_allV.append(np.nanargmin(s2_s[n_s < ntot]))
```
