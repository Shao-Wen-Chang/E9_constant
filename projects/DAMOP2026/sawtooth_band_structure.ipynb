```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameters
t = 1.0
V_offset = -2.0 * t

# Define k in the first Brillouin Zone: [-pi, pi]
k = np.linspace(-np.pi, np.pi, 200)

# 1. Analytical bands for the Normal Sawtooth
band_flat_normal = np.full_like(k, 2 * t)
band_disp_normal = -2 * t * (1 + np.cos(k))

# 2. Analytical bands for the Shifted Sawtooth
band_flat_shifted = band_flat_normal + V_offset
band_disp_shifted = band_disp_normal

# Set universal y-limits so both plots share the exact same scale
y_min = -4.5 * t
y_max = 2.5 * t

fig1, ax1 = plt.subplots(1, 1, figsize = (4, 3), dpi = 300)

ax1.plot(k / np.pi, band_flat_normal, color = "black", lw = 2, label = "Flat Band")
ax1.plot(k / np.pi, band_disp_normal, color = "black", lw = 2, label = "Dispersive Band")
# ax1.set_title("Normal Sawtooth")
# ax1.set_xlabel(r"$k / \pi$", size = 14)
# ax1.set_ylabel(r"Energy ($t$)")
ax1.set_xlim(-1, 1)
ax1.set_ylim(y_min, y_max)
ax1.tick_params(direction = "in", labelsize = 14)
ax1.set_xticks([-1, 0, 1])
ax1.set_yticks([-4, -2, 0, 2])

fig1.tight_layout()
```

```python
fig2, ax2 = plt.subplots(1, 1, figsize = (4, 3), dpi = 300)

ax2.plot(k / np.pi, band_flat_shifted, color = "black", lw = 2)
ax2.plot(k / np.pi, band_disp_shifted, color = "black", lw = 2)
# ax2.set_title("Shifted Sawtooth")
# ax2.set_xlabel(r"$k / \pi$")
# ax2.set_ylabel(r"Energy ($t$)")
ax2.set_xlim(-1, 1)
ax2.set_ylim(y_min, y_max)
ax2.tick_params(direction = "in", labelsize = 14)
ax2.set_xticks([-1, 0, 1])
ax2.set_yticks([-4, -2, 0, 2])

fig2.tight_layout()

plt.show()
```
